***DATA CLEANING***

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import streamlit as st
```

```python
df= pd.read_csv("data.csv")
df.head()
```

```python
# drop Date, Open message and language columns
df.drop(columns=["Date", "Open_Message", "Language"], inplace=True)
df.head()
```

```python
# check for missing values
df.isnull().sum()
```

```python
df.dtypes
```

```python
df.columns=['عدد أيام الغياب','آمن من التنمر','الجنس','المستوى الدراسي','نمط الغياب','جودة النوم ','الأسباب','غياب الأقران','الاستدراك','وسيلة النقل','سهولة الفهم']
```

```python
reason_for_absence =['عيان','مريض','مشكل فالعائلة','ماكنفهمش الدروس','خايف','مشكل فالنقل']
```

```python
# add new column for reason of absence with 1 in each column detected in reason column and 0 in the others
for reason in reason_for_absence:
    df[reason] = df['الأسباب'].apply(lambda x: 1 if reason in x else 0)
df.head()

```

```python
# drop the original reason column
df.drop(columns=['الأسباب'], inplace=True)
df.head()
```

```python
# same for the transport column
transport_means = ['المشي','النقل']
df['المشي']=df['وسيلة النقل'].apply(lambda x: 1 if 'على رجلي' in x else 0)
df['النقل']=df['وسيلة النقل'].apply(lambda x: 1 if 'النقل' in x else 0)
df.head()
```

```python
# drop the original transport column
df.drop(columns=['وسيلة النقل'], inplace=True)
df.head()
```

```python
# Standardize values

df["عدد أيام الغياب"] = df["عدد أيام الغياب"].replace({
    "0":0, "1-2":1.5, "3+":3
})

df["آمن من التنمر"] = df["آمن من التنمر"].replace({"إه":1, "لا":0})
df['جودة النوم ']  = df['جودة النوم '].replace({"إه":1, "لا":0})
df["الاستدراك"] = df["الاستدراك"].replace({"إيه":1, "لا":0})
df["سهولة الفهم"] = df["سهولة الفهم"].replace({"إه":1, "لا":0,"شي مرات":0.5})
df["غياب الأقران"] = df["غياب الأقران"].replace({"إه":1, "لا":0})
df["الجنس"] = df["الجنس"].replace({"ولد":1, "بنت":0})

```
